# Paso 1 : Importar librerias

```python
import pandas as pd
```

# Paso 2 : Importar datos

```python
admission = pd.read_csv('https://raw.githubusercontent.com/hermesherrera/python-class/refs/heads/main/Admission.csv')
```

```python
admission.head()
```

```python
admission.info()
```

```python
admission.describe()
```

# Paso 3 : Definir variable dependiente e independiente

```python
admission.columns
```

```python
y = admission['Chance of Admit ']
```

```python
X = admission.drop(['Serial No.','Chance of Admit '],axis=1)
```

# Paso 4 : Entrenamiento

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Paso 5 : Seleccionar modelo

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

# Paso 6 : Entrenamiento del modelo

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

# Paso 7 : Predicción del modelo

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

# Paso 8 : Precisión del modelo

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
